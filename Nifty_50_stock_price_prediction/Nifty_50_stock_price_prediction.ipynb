## - Importing Files and Libraries

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
Sc = StandardScaler()
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder
```

```python
# Creating the location of files
folder = 'C:/Users/Prakash Khatri/Downloads/stock_files'
```

```python
# creating a list of all the files 
files = os.listdir(folder)
```

```python
files
```

```python
# creating the absolute address of file with its name.
location = ['C:/Users/Prakash Khatri/Downloads/stock_files' + '/' + file for file in files]
```

```python
location
```

```python
# Cerating an empty list and appending all of the files in csv format
dfs = []
for col in range(len(files)):
    # code will read the location of the file into the programme
    df = pd.read_csv(location[col])
    # This code will add the column with dedicated file name that was just imported
    df['Name'] = files[col].replace(".NS.csv", "")
    # this code will switch the indexed postion of the name column
    df = df.iloc[:,[7,0,1,2,3,4,5,6]]
    # this code will append the files that were imported into a list
    dfs.append(df)
```

```python
dfs
```

```python
# This code will convert the data into Data Frame as a whole.
stock_df = pd.concat(dfs, ignore_index = True)
```

```python
stock_df
```

```python
# creating a copy of Original Data Frame.
df2 = stock_df.copy()
```

```python
# Extracting the numerical colunms
num = [cols for cols in stock_df.columns if stock_df[cols].dtypes != 'O']
```

```python
num
```

```python
# Outlier Analysis
for out in df2[num]:
    Quantile_1 = df2[out].quantile(0.25)
    Quantile_3 = df2[out].quantile(0.75)
    Inter_quartile_range = Quantile_3 - Quantile_1
    upper_limit = Quantile_3 + 1.5*Inter_quartile_range
    lower_limit = Quantile_1 - 1.5*Inter_quartile_range
    df2 = df2[(df2[out] <= upper_limit) & (df2[out] >= lower_limit)]
```

```python
# visualising Outliers
for cols in stock_df[num]:
    plt.figure()
    sns.boxplot(data=stock_df, y=cols)
```

```python
# so from the above plot we can see that amoung the diferent features "open" is the least correlated
# now we can build our model by taking 'Open' and 'Volumne' as the only feature.
df_data = df2.drop(['High', 'Low', 'Adj Close', 'Date'], axis= 1)
```

```python
new_num = ['Open', 'Volume']
# scaling and encoding the data
df_data[new_num] = Sc.fit_transform(df_data[new_num])
```

```python
# Using OnehotEncoder we'll encode the categorical colunms post then apped the data to create a new data frame
OHE = OneHotEncoder()
x_encoded = OHE.fit_transform(df_data[['Name']]).toarray()
```

```python
x_encoded
```

```python
# This is the final table for x which is to be used in train test split, post that Ml model creation
x = pd.concat([df_data.drop('Name', axis=1), pd.DataFrame(x_encoded)], axis =1)
```

```python
x
```

```python
x1 = x.dropna().reset_index(drop=True)
```

```python
x1
```

```python
# train test split
x = x1.drop('Close', axis= 1)
y = x1['Close']
x.columns = x.columns.astype(str)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# using Linear Ml model to determine the closing price
Lr.fit(X_train, y_train)
```

```python
y_pred = Lr.predict(X_test)
```

```python
y_pred
```

```python
# Using evaluation metrics to find out accuracy of the model

# mean Absolute error
print('mean Absolute error', metrics.mean_absolute_error(y_pred, y_test))
# mean Squared error      
print('mean Squared error', metrics.mean_squared_error(y_pred, y_test))
# root mean Squared erroe      
print('root mean Squared error', np.sqrt(metrics.mean_squared_error(y_pred, y_test)))
# r2 Score
print('R2 Score', metrics.r2_score(y_pred, y_test))
# Adjusted R2 score
print("Adjusted R2 Score", (1-((((1-metrics.r2_score(y_pred, y_test))*(df2.shape[0] -1))) / (df2.shape[0] - df2.shape[1] -1))))
```
